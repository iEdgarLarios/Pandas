<p>Matplotlib es una librería de Python que permite realizar gráficas 2D de excelente calidad. Es multiplataforma y puede ser usada desde scripts o desde la consola de Python.</p>

<p>Podemos definir el objetivo de matplotlib con las propias palabras de su creador John Hunter -- "matplotlib tries to make easy things easy and hard things possible" --</p>


<p>Matplotlib fue creado tomando como base a Matlab por lo que para las personas que han gráficado alguna vez en su vida con Matlab, Matplotlib les va a resultar muy familiar. Podemos exportar nuestras gráficas en los formatos de imágenes mas populares e incluso a formato latex para su inclusión en artículos científicos.</p>

<p>- ¿Si Matplotlib fue creado pensando en Matlab, porqué debería usarlo en vez de usar el mismo Matlab?</p>


<p>Existen varias razones. Es cierto que Matplotlib se baso en Matlab pero específicamente en su facilidad de uso. Matplotlib no trata de ser un reemplazo a Matlab como si lo es Octave. Matplotlib es un librería para Python que permite realizar gráficas brindando múltiples facilidades para lograr ese objetivo. Si queremos realizar gráficas de cualquier tipo y no necesitamos toda la cantidad de toolboxes que vienen con Matlab, matplotlib es justo lo que necesitamos.
Por otro lado, y no menos importante. Matplotlib es una librería libre lo cual quiere decir entre otras cosas que podemos usarla completamente gratis, lo cual es una ventaja de entrada frente a Matlab.</p>

```python
# Importando las librerías
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Preparando los datos
datos = np.arange(100,200)

plt.plot(datos,"b--")
plt.savefig('linea.jpg')
plt.show()
```

```python
datos2 = np.radians(datos)
datos2 = np.cos(datos)
```

```python
plt.plot(datos2)
plt.savefig('coseno.jpg')
plt.show()
```

```python
#https://problemsolvingwithpython.com/06-Plotting-with-Matplotlib/06.04-Saving-Plots/
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline


x = [0, 2, 4, 6]
y = [1, 3, 4, 8]

plt.plot(x,y)

plt.xlabel('x values')
plt.ylabel('y values')
plt.title('plotted x and y values')
plt.legend(['line 1'])

# save the figure
plt.savefig('plot.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Generación de un gráfico de Barras con matplotlib.pyplot
# 1.- Importar librerías 
import matplotlib.pyplot as plt
import numpy as np
# Crear un array para almacenar el primer conjunto de datos
x = np.array([5,7,9])
y = np.array([8,10,7])

# Crear un array para almacenar el segundo conjunto de datos 
x2 = np.array([6,8,10])
y2 = np.array([15,7,9])

# Crear las barras para el primer conjunto de datos 
plt.bar(x, y, align="center")

# Crear las barras para el segundo conjunto de datos
plt.bar(x2, y2, color = "g", align="center")

# Colocar un titulo a la grafica
plt.title("Grafica de barras")

#Incluir leyendas
plt.legend(["Conjunto 1","Conjunto 2"])

# por ultimo mostrar la grafica 
plt.show()
```

```python
# Generación de un gráfico de Paster con matplotlib.pyplot
# 1.- Importar librerías
import matplotlib.pyplot as plt
import numpy as np

# Crear un listas con datos a gráficar
manzanas = [10, 10, 25, 30]
nombres = ["Ana", "Juan", "Diana", "Catalina"]

# Crear la gráfica de pastel
plt.pie(manzanas, labels = nombres)

# Mostrar la gráfica
plt.show()
```

```python
# Mostrar los porcentajes de los valores 
plt.pie(manzanas, labels = nombres, autopct="%0.1f %%")
plt.show()
```

```python
# Personalizar colores 
colores = ["#EE6055","#60D394","#AAF683","#FFD97D"]
plt.pie(manzanas, labels = nombres, autopct="%0.1f %%", colors=colores)
plt.show()
```

```python
#Extrayendo una rebanada
desfase = (0, 0, 0, 0.1)
plt.pie(manzanas, labels = nombres, autopct="%0.1f %%", colors = colores, explode = desfase)
plt.show()
```
